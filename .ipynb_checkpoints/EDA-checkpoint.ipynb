# Exploratory data analysis

https://www.kaggle.com/datasets/rajyellow46/wine-quality

#### Our objective is to find quality of wine which is classification problem, we mark it from 1 to 9. Our data is inbalanced and even few quality marks are missing. We have 12 features  
#### 1.type: color of wine 
#### 2.fixed acidity: it corresponds to the set of low volatility organic acids such as malic, lactic, tartaric or citric acids and is inherent to the characteristics of the sample
#### 3.volatile acidity: it corresponds to the set of short chain organic acids that can be extracted from the sample by means of a distillation process: formic acid, acetic acid, propionic acid and butyric acid. It appears as a consequence of the metabolic transformations of the fruit and is zero or very close to it for fresh fruit
#### 4.citric acid: it is a weak organic acid, which is often used as a natural preservative or additive to drink to add a sour taste. It can also be used to neutralize surfaces that have been treated with basic substances
#### 5.residual sugar : it is the amount of sugar left in a wine when the alcoholic fermentation has been completed
#### 6.chlorides: the content of these elements is increased in wines coming from vineyards which are near the sea coast, which have brackish subâ€”soil or which have arid ground irrigated with salt water and the molar ratio cf Cl/Na+ therefore varies significantly and can even have a value close to one
#### 7.free sulfur dioxide : it aids in preventing microbial growth and the oxidation of wine
#### 8.total sulfur dioxide: it is the portion of SO2 that is free in the wine plus the portion that is bound to other chemicals in the wine such as aldehydes, pigments, or sugars
#### 9.density: it is density
#### 10.pH: it measures the acidity of the wine, determining its ripeness. A wine that has a low pH level (more acidic) typically tastes sour and dry, while a wine that is more alkaline (has a high pH) will taste flat and lack freshness due to increased bacterial growth
#### 11.sulphates: they help to maintain the freshness and flavour of wine and prolong its shelf life
#### 12.alchohol: it is percent of alchohol

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from scipy import stats
import os
import func
```

```python
data = pd.read_csv('winequalityN.csv')
```

#### Here we have imbalanced data

```python
spliter = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)

for train_index, test_index in spliter.split(data.iloc[:,:-1], data.iloc[:,-1]):
    for train_index, valid_index in spliter.split(data.iloc[train_index,:-1], data.iloc[train_index,-1]):
        train_data = data.iloc[train_index,:]
        train_data = train_data.reset_index(drop=True)
        
        validation_data = data.iloc[valid_index,:]
        validation_data = validation_data.reset_index(drop=True)
        
        test_data = data.iloc[test_index,:]
        test_data = test_data.reset_index(drop=True)
```

```python
def data_c(data):
    return data.iloc[:,-1:].value_counts(normalize=True).sort_index()
    
td = data_c(train_data)
vd = data_c(validation_data) 
ted = data_c(test_data)

pd.concat([td, vd, ted], axis=1, keys=['train_split', 'validation_split', 'test_split'])
```

```python
all_columns = list(train_data.columns)
numerical_columns = all_columns[1:-1]
categorical_columns = all_columns[0:1]
target = all_columns[-1]
```

## Handling missing data

```python
if train_data.isnull().sum().sum() == 0:
    print('you already handeled missing features!')
else:
    total_select = train_data.isnull().sum()
    
    total_select.plot(kind="bar", figsize = (6,4), fontsize = 10)
    
    plt.xlabel("Columns", fontsize = 20)
    plt.ylabel("Count", fontsize = 8)
    plt.title("Total Missing Values", fontsize = 20)    
```

#### We have few missing data, we handle it in next cell replacing missing data to mean of the traing data.

```python
for i in data.columns.tolist():
    if i in numerical_columns:
        mean = np.mean(train_data[i])
        train_data[i].fillna(value=mean, inplace=True)
        validation_data[i].fillna(value=mean, inplace=True)
        test_data[i].fillna(value=mean, inplace=True)
if train_data.isnull().sum().sum() != 0:
    print('you have missing values.')
else:
    print('done')
```

```python
type_target = train_data.groupby('type')[target].value_counts(normalize=True)
red = type_target['red'].sort_index()
white = type_target['white'].sort_index()
pd.concat([red, white, red/white], axis=1, keys=['red', 'white', 'red/white'])
```

## Visualizing data with histogram

```python
fig, ax = plt.subplots(len(numerical_columns)//2+1,2, figsize=(16,40))
ax = ax.flatten()
ax[-1].axis('off')
k=0

skew_data = np.abs(data[numerical_columns].skew())
skew_data[skew_data > 0.75]

for i in numerical_columns: 
    skew_data = np.abs(train_data.groupby('type')[i].skew())
    red = np.round(skew_data['red'], decimals=3)
    white = np.round(skew_data['white'], decimals=3)
    
    plot_type_num = sns.histplot(data=train_data[[i, categorical_columns[0]]], x=i, kde=True, hue=train_data[categorical_columns[0]], ax=ax[k])
    plot_type_num.set(xlabel=i + ': red ' + str(red) + '  white ' + str(white))
    if k == 0:
        plot_type_num.set_title('Wine')

    k = k + 1
```

## visualizing data with box

```python
len_numerical_columns = len(numerical_columns)
fig, axes = plt.subplots(len_numerical_columns//2+1,2,figsize=(20,30))
ax = axes.flatten()
ax[-1].set_axis_off()

for i, ax in enumerate(ax):
    if len_numerical_columns == i:
        continue
    z_score = stats.zscore(train_data[numerical_columns[i]])
    z_abs = (np.abs(z_score) > 3).sum()
    
    boxplot = sns.boxplot(x=train_data[numerical_columns].iloc[:,i],ax=ax)
    boxplot.set(xlabel=numerical_columns[i] + '; outliers: ' + str(z_abs))
```

#### In box plots we see outliers of features, additionaly I calculated count of outliers with z score.

## Feature engineering

```python
train_new_column = train_data['free sulfur dioxide'] / train_data['total sulfur dioxide']
train_new_column.rename('free/total sulfur', inplace=True)
valid_new_column = validation_data['free sulfur dioxide'] / validation_data['total sulfur dioxide']
valid_new_column.rename('free/total sulfur', inplace=True)
test_new_column = test_data['free sulfur dioxide'] / test_data['total sulfur dioxide']
test_new_column.rename('free/total sulfur', inplace=True)

train_data = pd.concat([train_new_column, train_data], axis=1)
validation_data = pd.concat([valid_new_column, validation_data], axis=1)
test_data = pd.concat([test_new_column, test_data], axis=1)
```

```python
numerical_columns.append('free/total sulfur')
```

```python
skew_data = np.abs(train_data.groupby('type')[numerical_columns[-1]].skew())
red = np.round(skew_data['red'], decimals=3)
white = np.round(skew_data['white'], decimals=3)

plot_type_num = sns.histplot(data=train_data[['free/total sulfur', categorical_columns[0]]], x='free/total sulfur', kde=True, hue=train_data[categorical_columns[0]])
plot_type_num.set(xlabel=numerical_columns[-1] + ': red ' + str(red) + '  white ' + str(white))
```

#### I added new feature which is division of free and total sulfur dioxides

```python
ohe_type = OneHotEncoder(sparse_output=False, drop='if_binary')
train_type = ohe_type.fit_transform(train_data[categorical_columns[0]].values.reshape(-1,1))

validation_type = ohe_type.transform(validation_data[categorical_columns[0]].values.reshape(-1,1))

test_type = ohe_type.transform(test_data[categorical_columns[0]].values.reshape(-1,1))

train_data_ohe = pd.concat([pd.DataFrame(train_type,columns=categorical_columns), train_data[numerical_columns+[target]]], axis=1)
validation_data_ohe = pd.concat([pd.DataFrame(validation_type,columns=categorical_columns), validation_data[numerical_columns+[target]]], axis=1)
test_data_ohe = pd.concat([pd.DataFrame(test_type,columns=categorical_columns), test_data[numerical_columns+[target]]], axis=1)
```
